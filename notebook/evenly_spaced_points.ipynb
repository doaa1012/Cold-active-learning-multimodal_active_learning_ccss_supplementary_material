```python
import os
import pandas as pd
import numpy as np

# --- CONFIG ---
dataset_path = DATASET_10311_Au_Pd_Pt_Rh_Ru_RT
init_indices = init_indices
target_col = "Resistance"
excluded_cols = ["ID", "x", "y"]
max_iter = 100

# --- OUTPUT DIR ---
dataset_name = os.path.splitext(os.path.basename(dataset_path))[0]
output_dir = os.path.join(EVENLY_PATH_Uncertainty, dataset_name)
os.makedirs(output_dir, exist_ok=True)

# --- LOAD DATA ---
df = pd.read_csv(dataset_path)
df[target_col] = np.log(df[target_col])

features = [c for c in df.columns if c not in excluded_cols + [target_col]]
target = [target_col]

# --- RUN ACTIVE LEARNING LOOP ---
mae_collection, mean_collection, stopping_index, gradient_stable_index = loop(
    df=df,
    features=features,
    target=target,
    init=init_indices,
    imax=max_iter,
    gp_class=GPBasic
)

# --- SAVE MAE ---
mae_df = pd.DataFrame({"Iteration": range(len(mae_collection)), "MAE": mae_collection})
mae_df.to_csv(os.path.join(output_dir, "mae_history.csv"), index=False)

# --- SAVE MEAN COLLECTION (each iteration predictions) ---
# --- SAVE MEAN COLLECTION (each iteration predictions) ---
mean_array = np.array([m.squeeze() for m in mean_collection])  # shape = (iterations, n_points)

# Transpose so rows = points, columns = iterations
mean_df = pd.DataFrame(mean_array.T)
mean_df.columns = [f"Iter_{i}" for i in range(mean_array.shape[0])]
mean_df.insert(0, "Index", range(mean_array.shape[1]))  # optional index column
mean_df.to_csv(os.path.join(output_dir, "mean_predictions_per_iteration.csv"), index=False)


# --- SAVE FINAL PREDICTIONS ---
final_pred = np.array(mean_collection[-1]).squeeze()  # from (342,1)  (342,)
final_true = np.exp(df[target_col].values).squeeze()

final_df = pd.DataFrame({
    "True_Resistance": final_true,
    "Predicted_Resistance": final_pred
})
final_df.to_csv(os.path.join(output_dir, "final_predictions.csv"), index=False)



# --- SAVE INFO ---
with open(os.path.join(output_dir, "info.txt"), "w") as f:
    f.write(f"Stopped at iteration: {stopping_index}\n")
    f.write(f"Gradient stable index: {gradient_stable_index}\n")
    f.write(f"Init indices: {init_indices}\n")

print(f"Results saved to: {output_dir}")
```
